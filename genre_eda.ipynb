```python
def concat_list(list_1, list_2):
    return eval(list_1) + eval(list_2)

train["genre"] = train.apply(lambda x :concat_list(x["title_to_genre"],x["tag_to_genre"]), axis=1)
```

```python
## songs에 얼마나 다양한 장르가 포함되는지 확인하자 (대분류기준)

train['plylst_genre_name'] = train["plylst_genre_name"].map(lambda x : eval(x))
```

```python
def unique_genre(genre_list):
    return list(set(genre_list))

def count_genre(genre_list):
    return len(genre_list)

train['unique_genre'] = train["plylst_genre_name"].map(lambda x :unique_genre(x))
train["genre_count"] = train["unique_genre"].map(lambda x : count_genre(x))
```
